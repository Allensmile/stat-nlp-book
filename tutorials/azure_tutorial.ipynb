## Virtual machine setup

Ok, now we have set up MS Azure, now we can build a virtual machine...

### Choosing a machine

Click on the **green + sign**, then **Compute**, and select **Ubuntu Server 16.04 LTS** as shown here:

![](../img/azure_tutorial/6.png)

You are also free to experiment with Windows Server 2012 if you wish, and remotely connect to it, but we won't do that in this tutorial.

On the following screen, select **Resource Manager** as deployment model and click on **Create**

![](../img/azure_tutorial/7.png)

On the following screen, configure basic settings:
- name of the virtual machine (e.g. stat-nlp-book)
- disk type (HDDs are cheaper)
- user name (remember this, we'll refer to this as USER_NAME)
- authentication type (**password** is easier, but go with **SSH public key** if you know what you're doing)
- set your password
- create a new resource group and name it (e.g. ucl-stuff)
- set location to 'West Central US' (this will reduce the total price of the machine and leave more cash for you to use. You're also free to chose another location)

![](../img/azure_tutorial/8.png)

...and click on OK to go to the next screen where you will need to pick your machine size. By clicking view all, you can see all the available machines. You will need a machine with at least 3GB of disk space.

![](../img/azure_tutorial/9.png)

Click ok until the Summary , and click OK there to create your machine. Wait until the machine is deployed. When it is deployed, you should be seeing a screen similar to the following one (note stat-nlp-book virtual machine running on the right side).

![](../img/azure_tutorial/10.png)

The machine is successfully up and running.
